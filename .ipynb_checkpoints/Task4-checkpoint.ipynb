## Countries_scraping :

```python
# Créer une liste pour stocker les noms de pays
countries_list = []
soup = BeautifulSoup(response.content, "html.parser")

# Trouver tous les éléments h2 qui contiennent les noms des pays
h2_elements = soup.find_all("h2")

# Parcourir les éléments h2
for h2 in h2_elements:
    # Récupérer le nom du pays
    country_name = h2.get_text()

    # Vérifier si le texte "[edit]" est présent
    if "[edit]" in country_name:
        # Supprimer le texte "[edit]" à partir du nom du pays
        country_name = country_name.replace("[edit]", "").strip()

    # Ajouter le nom du pays à la liste s'il n'est pas vide
    if country_name:
        countries_list.append(country_name)
```

```python
# Vérifier que la requête a réussi
if response.status_code == 200:
    # Utiliser BeautifulSoup pour analyser le contenu HTML
    soup = BeautifulSoup(response.content, "html.parser")

    # Créer une liste pour stocker les données des pays
    country_data = []

    # Trouver tous les éléments h2 qui contiennent les noms des pays
    h2_elements = soup.find_all("h2")

    # Parcourir les éléments h2
    for h2 in h2_elements:
        # Récupérer le nom du pays
        country_name = h2.get_text()

        # Trouver tous les éléments li associés au h2
        ul_element = h2.find_next_sibling(["ul", "ol"])
        if ul_element:
            li_elements = ul_element.find_all("li")

            # Vérifier s'il y a des éléments li
            if li_elements:
                # Créer un dictionnaire pour stocker les données du pays
                country_dict = {
                    "country_name": country_name,
                    "suffixes": []
                }

                # Parcourir les éléments li
                for li in li_elements:
                    # Récupérer le texte de l'élément li
                    sentence = li.get_text()

                    # Supprimer les éléments entre parenthèses
                    sentence = re.sub(r'\([^()]*\)', '', sentence)

                    # Vérifier s'il y a un ":" dans la phrase
                    if ":" in sentence:
                        # Extraire la partie avant le ":"
                        suffix = sentence.split(":")[0].strip()
                    else:
                        # Si aucun ":" n'est présent, prendre la phrase complète
                        suffix = sentence.strip()

                    # Ajouter le suffixe au dictionnaire du pays
                    country_dict["suffixes"].append(suffix)

                # Ajouter le dictionnaire du pays à la liste country_data
                country_data.append(country_dict)

    # Afficher les données des pays
    for country in country_data:
        country_name = country["country_name"]
        # Supprimer le texte "[edit]" de la valeur country_name
        country_name = country_name.replace("[edit]", "").strip()

        print("Nom du pays:", country_name)
        print("Suffixes:", country["suffixes"])
        print("-------------")

else:
    print("Erreur lors de la requête à la page.")
```

```python
# Faire une requête GET à la page
response = requests.get(url)

# Vérifier que la requête a réussi
if response.status_code == 200:
    # Utiliser BeautifulSoup pour analyser le contenu HTML
    soup = BeautifulSoup(response.content, "html.parser")

    # Créer une liste pour stocker les données des pays
    country_data = []

    # Trouver tous les éléments h2 qui contiennent les noms des pays
    h2_elements = soup.find_all("h2")

    # Parcourir les éléments h2
    for h2 in h2_elements:
        # Récupérer le nom du pays
        country_name = h2.get_text()

        # Supprimer "[edit]" qui suit le nom du pays
        country_name = re.sub(r"\[edit\]$", "", country_name).strip()

        # Trouver tous les éléments li associés au h2
        ul_element = h2.find_next_sibling(["ul", "ol"])
        if ul_element:
            li_elements = ul_element.find_all("li")

            # Vérifier s'il y a des éléments li
            if li_elements:
                # Créer un dictionnaire pour stocker les données du pays
                country_dict = {
                    "Country": country_name,
                    "Company_Suffixes": []
                }

                # Parcourir les éléments li
                for li in li_elements:
                    # Récupérer le texte de l'élément li
                    sentence = li.get_text()

                    # Supprimer les éléments entre parenthèses
                    sentence = re.sub(r'\([^()]*\)', '', sentence)

                    # Supprimer les occurrences de [edit]
                    sentence = re.sub(r'\[edit\]', '', sentence)

                    # Vérifier s'il y a un ":" dans la phrase
                    if ":" in sentence:
                        # Extraire la partie avant le ":"
                        suffix = sentence.split(":")[0].strip()
                    else:
                        # Si aucun ":" n'est présent, prendre la phrase complète
                        suffix = sentence.strip()

                    # Ajouter le suffixe au dictionnaire du pays
                    country_dict["Company_Suffixes"].append(suffix)

                # Ajouter le dictionnaire du pays à la liste country_data
                country_data.append(country_dict)

    # Créer un DataFrame à partir de country_data
    df = pd.DataFrame(country_data)
    df=df.head(66)
    # Enregistrer le DataFrame dans un fichier CSV
    df.to_csv("country_data.csv", index=False)

    print("Les données ont été enregistrées dans country_data.csv.")

else:
    print("Erreur lors de la requête à la page.")
```

### The other countries (table) : 

```python
import requests
from bs4 import BeautifulSoup

# Définir l'URL de la page à scraper

# Envoyer une requête GET pour obtenir le contenu de la page
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Utiliser BeautifulSoup pour analyser le contenu HTML de la page
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Trouver la balise h4 avec le texte "Others"
    h4_tags = soup.find_all("h4")
    for h4_tag in h4_tags:
        span_tag = h4_tag.find("span", text="Others")
        if span_tag:
            # Trouver le tableau suivant la balise h4
            table = h4_tag.find_next("table")
            
            # Vérifier si le tableau a été trouvé
            if table:
                country_list = []  # Liste pour stocker les noms des pays
                
                # Parcourir les lignes du tableau
                for row in table.find_all("tr"):
                    # Parcourir les cellules de chaque ligne
                    cells = row.find_all("td")
                    if len(cells) > 0:
                        # Ajouter le contenu de la première cellule à la liste
                        country_name = cells[0].get_text(strip=True)
                        country_list.append(country_name)
                        
                        # Afficher le contenu de chaque cellule
                        for cell in cells:
                            print(cell.get_text(strip=True))
                        print("--------------------")
                
                # Afficher la liste des noms de pays
                country_list.remove('State, territory or commonwealth')
                print("List of countries:", country_list)
            else:
                print("Table not found.")
            
            # Sortir de la boucle une fois le tableau trouvé
            break
    else:
        print("Section 'Others' not found.")
else:
    print("Failed to retrieve the webpage.")
```

```python
# Charger le fichier CSV avec pandas
df = pd.read_csv('data0.csv')

# Expression régulière pour trouver les mots entre guillemets
pattern = r'"([^"]*)"'

# Fonction pour extraire les mots entre guillemets d'une phrase
def extraire_mots_entre_guillemets(phrase):
    return re.findall(pattern, phrase)

# Appliquer la fonction sur la colonne 'texte' et créer une nouvelle colonne 'mots_guillemets'
df['Company_Suffixes'] = df['Text'].apply(extraire_mots_entre_guillemets)

df = df.drop('Text', axis=1)

df = df.drop(df.index[0])

# Afficher le résultat
df.to_csv("data.csv", index=False)

df
```

# Correction :

```python
import pandas as pd


# Créer un dictionnaire pour les suffixes de chaque pays
suffixes = {
    'Georgia': ['corporation', 'incorporated', 'company', 'limited', 'corp.', 'inc.', 'co.', 'ltd.'],
    'Nebraska': ['corporation', 'incorporated', 'company', 'limited', 'corp.', 'inc.', 'co.', 'ltd.'],
    'Pennsylvania': ['Corporation', 'Corp.', 'Company', 'Co.', 'Incorporated', 'Inc.', 'Limited', 'Ltd.', 'Association.', 'Fund.', 'Syndicate'],
    'Puerto Rico': ['Corporation', 'Corp.', 'Inc.'],
    'South Dakota': ['corporation', 'incorporated', 'company', 'limited', 'corp.', 'inc.', 'co.', 'ltd.']
}

# Itérer sur chaque ligne du DataFrame
for i, row in df.iterrows():
    country = row['Country']
    if country in suffixes:
        df.at[i, 'Company_Suffixes'] = ', '.join(suffixes[country])

# Vérifiez le DataFrame pour voir si les modifications ont été apportées correctement

df.to_csv("data.csv", index=False)
df
```
